## Vocabulary: Module vs Package

- **Module**: one `.py` file (e.g., `profiling.py`)
- **Package**: a folder of modules (e.g., `csv_profiler/` with `__init__.py`)

**Why packages?**
- Organize code by responsibility
- Reuse code across scripts
- Easier testing and maintenance


## What happens when you `import something`?

Python searches for `something` in this order:
1. Built-in modules
2. Installed packages (your environment)
3. Your project paths (current folder + `sys.path`)


```python
import sys

print("sys.path (where Python looks for imports):")
for p in sys.path:
    print(" -", p)
```

## PYTHONPATH: Add folders to import search

If your code lives in `src/`, add it to the path:
- Unix/macOS: `PYTHONPATH=src uv run python main.py`
- Windows PowerShell: `$env:PYTHONPATH="src"; uv run python main.py`

```python
# Test for: slugify function
def test_exercise():
    """Run tests with detailed feedback."""
    errors = []
    
    # Test 1: Basic functionality
    try:
        result = slugify("My Report 01")
        if result != "my-report-01":
            errors.append(f"Expected 'my-report-01', got '{result}'")
    except Exception as e:
        errors.append(f"Function raised an error: {e}")
    
    # Test 2: With extra whitespace
    try:
        result = slugify("  Hello World  ")
        if result != "hello-world":
            errors.append(f"Expected 'hello-world', got '{result}' (whitespace not handled correctly)")
    except Exception as e:
        errors.append(f"Function raised an error with whitespace: {e}")
    
    # Test 3: Single word
    try:
        result = slugify("Python")
        if result != "python":
            errors.append(f"Expected 'python', got '{result}' (single word not handled correctly)")
    except Exception as e:
        errors.append(f"Function raised an error with single word: {e}")
    
    # Test 4: Return type
    try:
        result = slugify("Test")
        if not isinstance(result, str):
            errors.append(f"Function should return a string, got {type(result).__name__}")
    except Exception as e:
        errors.append(f"Function raised an error: {e}")
    
    if errors:
        print("âŒ Some tests failed. Here's what went wrong:\n")
        for i, error in enumerate(errors, 1):
            print(f"{i}. {error}")
        raise AssertionError(f"{len(errors)} test(s) failed")
    else:
        print("âœ… All tests passed! Great job!")

test_exercise()
```

```python
# Solution
def slugify(text: str) -> str:
    """Turn 'Report Name' â†’ 'report-name'."""
    cleaned = text.strip().casefold()
    parts = cleaned.split()
    return "-".join(parts)
```

```python
import time

start = time.perf_counter_ns()
# Simulate some work
time.sleep(0.001)  # 1ms
end = time.perf_counter_ns()

elapsed_ms = (end - start) / 1_000_000
print(f"Elapsed: {elapsed_ms:.2f}ms")
```
